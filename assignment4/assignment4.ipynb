```python
import pandas as pd
pf = pd.read_csv('./panel_fin.csv')
#pf = pf[pf['Quarter']>80]
pf[pf['l_ret']=='.']['l_ret'] = 0.0
pf[['l_ret']] = pf[['l_ret']].astype('float')
pf.to_csv('panel_fin.csv')
pf = pf.set_index(['Entity', 'Quarter'])
pf
```

```python
from linearmodels.panel import PooledOLS
from linearmodels.panel import PooledOLS
import statsmodels.api as sm
pf = pd.read_csv('./panel_fin.csv')
pf = pf.set_index(['Entity', 'Quarter'])

exog_vars = ['l_ret','l_btm','l_e_p']
exog = sm.add_constant(pf[exog_vars])
mod = PooledOLS(pf.ret, exog)
pooled_res = mod.fit()
print(pooled_res)
```
